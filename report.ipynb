```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/clean_items.csv", usecols=["score"])
df["score"] = pd.to_numeric(df["score"], errors="coerce")
scores = df["score"].dropna()
plt.figure(figsize=(6,4))
plt.hist(scores, bins=20, edgecolor="black")
plt.xlabel("Official Score")
plt.ylabel("Count")
plt.title("Distribution of Anime Scores")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from IPython.display import Markdown

train = pd.read_csv("data/ratings_train.csv",
                    usecols=["user_id", "anime_id"])

stats = {
    "Users":   train.user_id.nunique(),
    "Items":   train.anime_id.nunique(),
    "Ratings": len(train),
    "Sparsity": f"{len(train)/(train.user_id.nunique()*train.anime_id.nunique()):.4%}"
}

md_table = pd.DataFrame([stats]).to_markdown(index=False)

Markdown(md_table)
```
